```python
import requests
import json

# Reading the destinations file.
my_text_opened = open('dests.txt', 'r', encoding='utf-8')
my_text_to_read = my_text_opened.read()
my_destinations_list = my_text_to_read.split('\n')

# Defining some variables to access the APIs and save the data.
params_dict = dict()
my_api_key = # Go to my api key in the submission folder.'AIzaSyA8WFU8HD1TG-T0a0VZgxLzX4520XtDJKs'
params_dict["key"] = my_api_key
basic_geocode_url="https://maps.googleapis.com/maps/api/geocode/json"
my_TLV_dstnc_matrix_list = []
my_clear_dict_list = []

# Defining my help function.
def distances_info(destination):
    basic_dist_url = "https://maps.googleapis.com/maps/api/distancematrix/json?units=imperial&origins=Tel Aviv-Yafo"
    dest_str = "&destinations="+destination
    API_key_str = "&key="+my_api_key
    dist_mat_url = basic_dist_url+dest_str+API_key_str
    dist_response = requests.get(dist_mat_url).json()
    dist_text=dist_response['rows'][0]['elements'][0]['distance']['text']
    only_dist_in_miles=dist_text[:len(dist_text)-3]
    only_digits = only_dist_in_miles.replace(',','')
    to_float = float(only_digits)
    kilometers_distance = round(to_float*1.6, 1)
    travel_time_txt = dist_response['rows'][0]['elements'][0]['duration']['text']
    return kilometers_distance, travel_time_txt

# Activating my program.
for dest in my_destinations_list:
    dest_dict = dict()
    params_dict["address"] = dest
    #params_dict = {"key":my_api_key, "address":my_address}
    response = requests.get(basic_geocode_url, params=params_dict).json()
    dest_name = response['results'][0]['address_components'][0]['long_name']
    lat_n_lng_dict = response['results'][0]['geometry']['location']
    lat, lng = lat_n_lng_dict['lat'],lat_n_lng_dict['lng']
    distance, travel_time = distances_info(dest_name)
    dest_dict[dest_name] = (distance, travel_time, lng, lat)
    my_TLV_dstnc_matrix_list.append(dest_dict)
    
    clear_dict = {"destination":dest_name, "km_distance":distance,"travel_time":travel_time, "longitude":lng, "latitude":lat}
    my_clear_dict_list.append(clear_dict)

# All the data, exactly as requested.
print(my_TLV_dstnc_matrix_list)

# Printing the data in a clearer format.
json_dict = json.dumps(my_clear_dict_list, indent=4, ensure_ascii=False)
print('')
print("All the data below is about the locations and distances of some cities from Tel-Aviv:")
print(json_dict)
```
